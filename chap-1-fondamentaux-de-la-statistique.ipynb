## Moyenne et médiane

Avoir une valeur unique qui résume une situation.

### La moyenne

La moyenne est très sensible aux valeurs extrêmes.

$\boxed{\bar{x} = \frac{1}{n} \sum_{i=1}^{n} x_{i}}$

### La médiane

La médiane n'est pas sensible aux valeurs extrêmes.

**Nombre de valeurs paires**

Tri croissant des valeurs :

$\boxed{3, 4, 5, 5, 9, 12, 12, 13, 20}$

Le nombre de valeurs étant impair la médiane est donc la valeur au milieu de la série 

$\boxed{m_e := q_{1/2} = \textbf{9}}$

**Nombre de valeurs impaires**

$\boxed{3, 4, 5, 9, 12, 12, 13, 20}$

Le nombre de valeurs étant impair la médiane est donc la valeur au milieu de 9 et 12 soit 10,5. En réalité, nous aurions pu choisir n'importe quel nombre entre 9 et 12 mais en général on prend la valeur moyenne entre ces 2 nombres.

$\boxed{m_e := q_{1/2} = \textbf{10,5}}$

## Les quantiles

Ils permettent de se représenter un jeu de données.

- **Médiane**
    - 50% des valeurs ≤ médiane
    - 50% des valeurs ≥ médiane
- **1er quartile**
    - 25% des valeurs ≤ q1
    - 75% des valeurs ≥ q1
- **3ème quartile**
    - 75% des valeurs ≤ q3
    - 25% des valeurs ≥ q3
    

### **Exemple échantillon :**

$\boxed{3, 4, 5, 5, 9, 12, 12, 13, 20}$

***p*** : quartile, ex : 1er quartile = 1/4

***n*** : taille de l'échantillon

$\boxed{Q=p(n+1)}$

$\boxed{Q_\frac{1}{4}=\frac{1}{4}\left(9+1\right)}$

$\boxed{Q_\frac{1}{4}=\frac{10}{4}=\textbf{2,5}}$ 

Ainsi, toutes les valeurs entre 0 et la 2,5ème valeur font partie du 1er quartile, pas grave si ce n'est pas une valeur entière.

## Variance et écart-type

Mesure la dispersion des données par rapport à la moyenne. Plus exactement, la somme des écarts par rapport à la moyenne.

### La variance

$\boxed{V = \frac{1}{n} \sum_{i=1}^{n} (x_{i}-\bar{x})^2}$ ou $\boxed{\sigma^2 = \frac{1}{n} \sum_{i=1}^{n} (x_{i}-\bar{x})^2}$ 

### L'écart-type

$\boxed{\sigma = \sqrt{\frac{1}{n} \sum_{i=1}^{n} (x_{i}-\bar{x})^2}}$

Soit : $\boxed{\sigma = \sqrt{V}}$

## La covariance

Dispersion conjointe de 2 variables. Est-ce que quand la première variable progresse, est-ce que la seconde évolue dans le même sens ? La covariance permet de calculer cette dispersion.

$\boxed{\sigma_{xy} = \frac{1}{n} \sum_{i=1}^{n} (x_{i}-\bar{x}) (y_{i}-\bar{y})}$

## Coefficient de corrélation de Pearson

Le coefficient de corrélation c'est le calcul de la covariance sur des variables transformées. La transformation est la division par l'écart-type de la variable concernée.

- Ainsi, on calcule l'écart type pour chaque variable :

$\sigma_{duree} = 33,5$

$\sigma_{retard} = 5,6$

- On divise chaque observation par son écart-type :

| Durée | Retard |
| --- | --- |
| 2,97 | 3,06 |
| 4,67 | 2,52 |
| 5,58 | 5,22 |
| 3,07 | 2,34 |
| 3,07 | 2,34 |
| 3,23 | 3,06 |

Si on recalcule l'écart-type des 2 variables, on tombe sur un écart-type de 1 pour chaque variable.

- En reprenant la formule de la covariance, le résultat est ***0,76***.

Une corrélation positive indique une croissance des valeurs dans la même direction, a contrario, une corrélation négative indique une croissance d'une variable et une décroissance de l'autre variable.

Le coefficient de corrélation varie entre -1 et 1.

### Les pièges de la corrélation

Corrélation ne signifie pas causalité.

Les indicateurs de corrélation ou de dépendance ne donnent pas de sens à la corrélation.

Statistiquement, aucun indicateur permet de déterminer le sens de la corrélation.

**Exemple de piège :**

En été, il existe une corrélation entre les ventes de glace et les coups de soleil. Il est évident que cette corrélation ne représente pas une causalité.

La corrélation fonctionne pour des variables de type linéaire. En effet, si vous prenez y= x^2, alors vous trouverez une corrélation de 0 car la parabole a une symétrie parfaite.
