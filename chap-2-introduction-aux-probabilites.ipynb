## Le th√©or√®me de Bayes

$\boxed{{P(A\cap B)}=P(A)\times P(B|A)}$

$\boxed{{P(A\cap B)}=P(B)\times P(A|B)}$ , soit : $\boxed{P(A)\times P(B|A)=P(B)\times P(A|B)}$

$\boxed{\frac{P(A)\times P(B|A)}{P(B)}=\frac{P(B)\times P(A|B)}{P(B)}}$

**La formule du th√©or√®me de Bayes**

Nous supprimons $P(B)$ sur l'√©quation de droite et passons $P(A|B)$ √† gauche

$\boxed{P(A|B)=\frac{P(A)\times P(B|A)}{P(B)}}$

$\boxed{P(A|B)=\frac{P(B|A)\times P(A)}{(P(B|A)\times P(A)) + (P(B|\bar{A})\times P(\bar{A}))}}$

**Exemple :** Se faire d√©pister dans un laboratoire pour une maladie.

- Si le patient n'est pas malade, le test est n√©gatif dans 95% du temps ‚áí $\boxed{P(\bar{B}|\bar{A})=0,95}$
